```python
import pandas as pd
pd.set_option('display.max_rows', None)  # Show all rows
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.width', 1000)  # Set display width


incomplete_tsv = '/groups/doudna/projects/kenneth/mcpol/taxa/mmseqs_0808.defense.annotated.locus.poi.dedup.taxID.tsv'
complete_tsv = '/groups/doudna/projects/kenneth/mcpol/taxa/mmseqs_0805.defense.annotated.locus.poi.dedup.fixed.tsv'

# Load the files
file1 = pd.read_csv(incomplete_tsv, sep='\t', dtype={'TaxID': str})  # The file with "Error" rows
file2 = pd.read_csv(complete_tsv, sep='\t', dtype={'TaxID': str})    # The file with full data

# Merge the two files on ProteinID
merged = file1.merge(file2, on='ProteinID', how='left', suffixes=('_incomplete', '_complete'))

# Fill missing values in the incomplete file with data from the complete file
# merged['ScientificName'] = merged['ScientificName_incomplete'].fillna(merged['ScientificName_complete'])
merged['ScientificName'] = merged['ScientificName_complete']
# merged['TaxID'] = merged['TaxID_incomplete'].fillna(merged['TaxID_complete'])
merged['TaxID'] = merged['TaxID_complete']

print(len(merged))
# merged[merged['ScientificName_incomplete'] == 'Error']

# Drop the extra columns
result = merged[['ProteinID', 'ScientificName', 'TaxID']]
result.head()

# Save the result
# result.to_csv('/groups/doudna/projects/kenneth/mcpol/taxa/mmseqs_0808.defense.annotated.locus.poi.dedup.fixed.tsv', sep='\t', index=False)
```

```python
accessions_filepath = '/groups/doudna/projects/yoon/cas10/mcpol/neighbor2/mmseqs_0808.defense.annotated.locus.csv'
accessions = pd.read_csv(accessions_filepath)
accessions = accessions.drop(['Protein ID', 'Translation'], axis=1)
accessions = accessions.drop_duplicates(subset='Protein of Interest', keep='first')
accessions = accessions.rename(columns={'Locus Tag': 'accession_id', 'Protein of Interest': 'Protein of Interest'})
print(accessions.shape)
accessions.head()
```

```python
mmseqs_0505_filepath = '/groups/doudna/projects/kenneth/mcpol/taxa/mmseqs_0505.defense.annotated.locus.poi.dedup.final.tsv'
mmseqs_0505 = pd.read_csv(mmseqs_0505_filepath, sep='\t', dtype={'TaxID': str})
mmseqs_0505 = mmseqs_0505.rename(columns={'ProteinID': 'Protein of Interest'})

# Rename the ProteinID column to Protein of Interest
mmseqs_0505.columns.values[0] = 'Protein of Interest'
print(mmseqs_0505.shape)
print(accessions.shape)
# Remove trailing spaces from 'Protein of Interest' in mmseqs_0505
print(mmseqs_0505['Protein of Interest'])
mmseqs_0505['Protein of Interest'] = mmseqs_0505['Protein of Interest'].str.strip()
# Merge based on Protein of Interest
mmseqs_0505_merge = mmseqs_0505.merge(accessions, how='left', on='Protein of Interest')

# Reorder the DataFrame columns
cols = mmseqs_0505_merge.columns.tolist()
# Move the last column to the first position
cols = [cols[-1]] + cols[:-1]
mmseqs_0505_merge = mmseqs_0505_merge[cols]

# Save the merged file
mmseqs_0505_merge.to_csv('/groups/doudna/projects/kenneth/mcpol/taxa/mmseqs_0505.defense.annotated.locus.poi.dedup.final.tsv', sep='\t', index=False)

print(mmseqs_0505_merge.shape)
mmseqs_0505_merge.head()
```
